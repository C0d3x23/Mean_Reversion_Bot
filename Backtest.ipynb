```python
import pandas as pd
from backtesting import Strategy, Backtest
from backtesting.lib import resample_apply, crossover
import talib as ta
import warnings

data = pd.read_csv('ETHUSDT_5m_2019-12-01-2023-05-01_futures.csv', index_col='Date', parse_dates=True)

class MeanReversion(Strategy):
    m15_rsi = 2
    m15_sma = 200
    d_sma = 200
    upper_rsi = 90
    lower_rsi = 10

    def trading_session_data(self):
        self.start_time = pd.to_datetime('08:00:00').time()
        self.end_time = pd.to_datetime('20:00:00').time()
        return self.start_time <= self.data.index[-1].time() <= self.end_time
            
    def init(self):

        self.rsi = self.I(ta.RSI, self.data.Close, self.m15_rsi)

        self.sma_m15 = self.I(ta.SMA, self.data.Close, self.m15_sma)
        
        self.d_trend = resample_apply(
            'D', ta.SMA, self.data.Close, self.d_sma)                                                        

    def next(self):

        trading_session = self.trading_session_data()

    # Check if current time is outside of 8am - 8pm time range and there's an open position
        if not trading_session and self.position:
            self.position.close()
            return

        price = self.data.Close[-1]

        if trading_session:
            if price > self.d_trend[-1] and not self.position:
                if self.rsi[-1] < self.lower_rsi:
                    self.buy()

        elif (self.rsi[-1] > self.upper_rsi):
            self.position.close()


backtest = Backtest(data, MeanReversion, cash=100000, commission=.001, exclusive_orders=True)
stats = backtest.run()
print(stats)
```
